Dataset Loading

```python
import pandas as pd

# Load the dataset
file_path_hles = '../../DS2/Dataset/SurveyData/SurveyData_CSV_DAP_2021_HLES_dog_owner_v1.0.csv'
file_path_afus = '../../DS2/Dataset/SurveyData/SurveyData_CSV_DAP_2021_AFUS_dog_owner_v1.0.csv'
dog_hles = pd.read_csv(file_path_hles, delimiter=',')  # Assuming tab-delimited format based on the file extension
dog_afus = pd.read_csv(file_path_afus, delimiter=',')
```

Show figure and count

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing the age distribution of the dogs
sex = dog_hles['dd_sex'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=sex.index, y=sex.values)
plt.title('Sex Distribution of Dogs')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

sex_counts = dog_hles['dd_sex'].value_counts()

print("1= Male, 2= Female")
print(sex_counts)
```

One-hot Encode

```python
#encode
one_hot_encoded_sex = pd.get_dummies(dog_hles['dd_sex'])

print(one_hot_encoded_sex)

# important to do this!Please keep the dog_id in combined_data, easy for checking
combined_data = pd.concat([dog_hles[['dog_id']], one_hot_encoded_sex], axis=1)
# make a new .csv file can load by our model.ipynb
combined_data.to_csv('one_hot_encoded_sex_with_id.csv', index=False)
```
